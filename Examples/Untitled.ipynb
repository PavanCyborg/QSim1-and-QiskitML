```python
from qiskit.primitives import BackendSampler, BackendEstimator
from qiskit_aer import Aer
backend=Aer.get_backend("qasm_simulator")
backend_s = BackendSampler(backend=backend,options={'default_shots':1})
backend_e = BackendEstimator(backend=Aer.get_backend("dm_simulator"),options={'default_shots':1})
```

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile

# Define 4-qubit quantum register and 4-bit classical register
qreg = QuantumRegister(4, "q")
creg = ClassicalRegister(4, "c")

# Create circuit with both registers
qc = QuantumCircuit(qreg, creg)

# Apply some gates
qc.h(qreg[0])           # Hadamard on qubit 0
qc.x(qreg[1])           # Pauli-X on qubit 1
qc.cx(qreg[0], qreg[2]) # CNOT q0 -> q2
qc.z(qreg[3])           # Pauli-Z on qubit 3

# Manually add measurement for each qubit
qc.measure(qreg[0], creg[0])
qc.measure(qreg[1], creg[1])
qc.measure(qreg[2], creg[2])
qc.measure(qreg[3], creg[3])

# Show the circuit
print(qc.draw())
```

```python
# Transpile circuit for this backend
tqc = transpile(qc, backend)
```

```python
# Run on dm_simulator
job = backend.run(tqc,shots=1000)
result = job.result()

```
